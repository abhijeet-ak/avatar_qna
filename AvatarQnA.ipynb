```python
from ipywebrtc import AudioRecorder, CameraStream
from IPython.display import Audio
from pipeline import QnAAvatar
import warnings
warnings.filterwarnings('ignore')
```

```python
camera = CameraStream(constraints={'audio':True, 'video':False})
avatar_ear = AudioRecorder(stream=camera)
```

### Intializing Digital Avatar

```python
avatar = QnAAvatar('sample.json')
avatar.hear = avatar_ear 
```

### Knowledge Base
Below we have a content for the Avatar. Avatar will hold this information with itself.<br>
And based on the user query we want avatar to be able to answer any user queries regarding the content. 

```python
avatar.get_context()
```

```python
avatar.hear
```

###  Convert the sound query to text

```python
transcribed_text = avatar.speechToText(['q1.wav'])
transcribed_text
```

### Figuring out anwser of the question

```python
ans, _ = avatar.answer_quesion(transcribed_text)
ans
```

### Text to Speech

```python
audio = avatar.speak_answer(ans)[0]
Audio(audio, rate=22050)
```

### Whole Picture

```python
ans = avatar.execute_pipeline(['q1.wav'])
Audio(ans, rate=22050)
```
